```python
import torch
import torch.nn as nn

# 1. Check PyTorch version and GPU availability
print("PyTorch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("CUDA device count:", torch.cuda.device_count())
if torch.cuda.is_available():
    print("Current CUDA device:", torch.cuda.get_device_name(torch.cuda.current_device()))

# 2. Create a simple tensor and move to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
x = torch.randn(2, 3).to(device)
print("\nRandom tensor x:\n", x)

# 3. Define a simple linear layer and pass tensor through it
linear = nn.Linear(3, 2).to(device)
output = linear(x)
print("\nLinear layer output:\n", output)

# 4. Perform a simple loss computation
target = torch.randn(2, 2).to(device)
loss_fn = nn.MSELoss()
loss = loss_fn(output, target)
print("\nLoss:", loss.item())

# 5. Backward pass
loss.backward()
print("\nBackward pass completed successfully!")
```
