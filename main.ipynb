##  tight_layout   axis('off')

 要减小子图（axes）之间的水平和垂直间距，您需要调整 plt.subplots_adjust() 函数中的 wspace 和 hspace 参数。



wspace 的单位解析 📊

wspace (Width Space) 并不是一个绝对的长度单位（如像素或英寸），而是 Matplotlib 中用于控制子图之间水平间距的一个相对比例值。

核心定义

wspace 的值代表着子图之间保留的水平间距，表示为 平均子图宽度 的一个分数（Fraction）。

$$\text{水平间距} = \text{wspace} \times \text{平均子图宽度}$$

值的含义举例

wspace 值含义说明视觉效果0.0水平间距为 0，子图应紧密相连（理想情况下）。最紧凑，子图之间几乎没有空隙。0.2 (默认值)水平间距为平均子图宽度的 $20\%$。保持一定的间距，避免内容过于拥挤。1.0水平间距等于一个子图的平均宽度。间距较大。

# MLP layer implementation

```python
import paddle 
import paddle.nn as nn 

class MLP(nn.Layer):
    def __init__(self,
                inputDimension,
                outputDimension,                
                mlpRatio =4,
                dropoutRate = 0):
        super(MLP,self).__init__()
        self.inputDimension = inputDimension
        self.outputDimension = outputDimension
        self.mlpRatio = mlpRatio
        self.dropoutRate = dropoutRate
        self.activeFunction = nn.GELU()
        self.hiddenDimension = int(self.inputDimension*self.mlpRatio)
        self.fc1 = nn.Linear(in_features= self.inputDimension,out_features= self.hiddenDimension)
        self.fc2 = nn.Linear(in_features=self.hiddenDimension,out_features = self.outputDimension)
        self.dropout = nn.Dropout(p = self.dropoutRate)
    def forward(self,inputs):
        x = self.fc1(inputs)
        x = self.activeFunction(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.dropout(x)
        return x





testX = paddle.randn((1,197,768))


VitMLP = MLP(inputDimension= 768,outputDimension=768,mlpRatio=4,dropoutRate=0.0)
testX = VitMLP(testX)
print(testX.shape)

print(testX)
        
```

# patch embedding implementation

```python
import paddle 
import paddle.nn as nn


class PatchEmbedding(nn.Layer):
    def __init__(self,imageSize = 224,inChannels = 3,patchSize = 16,embeddingDimension = 768):
        super(PatchEmbedding,self).__init__()
        self.imageSize = imageSize 
        self.inChannels = inChannels
        self.patchSize = patchSize
        self.embeddingDimension = embeddingDimension 
        self.patchNum = (self.imageSize//self.patchSize)**2
        self.linearProjection = nn.Linear(in_features=self.patchSize*self.patchSize*self.inChannels,out_features=self.embeddingDimension)
        # W_linearproj->768*768
        self.positionEmbedding = self.create_parameter(shape=[1,self.patchNum+1,self.embeddingDimension],dtype='float32')
        #1,N+1,D
        self.classEmbedding = self.create_parameter(shape= [1,1,self.embeddingDimension], dtype = 'float32')
        #1,1,D



    def forward(self,inputs):
        B,C,H,W = inputs.shape
        print(B,C,H,W)
        Hp = H//self.patchSize
        Wp = W//self.patchSize
        P = self.patchSize
        x = inputs.transpose(perm=[0,2,3,1])
        # B,H,W,C
        x = x.reshape(B,Hp,P,Wp,P,C)
        print(x.shape)
        # B,Hp,P,Wp,P,C 
        x = x.transpose(perm=[0,1,3,2,4,5])
        # B ,Hp,Wp,P,P,C -> B,14,14,16,16,3
        print(x.shape)
        x = x.reshape(shape = [B,Hp,Wp,P*P*C])
        #B, H//P,W//P,P*P*C ->B,14,14,768
        print(x.shape)
        x = x.reshape(shape = [B,Hp*Wp,P*P*C])
        # B,Hp*Wp,P*P*C ->B,196,768
        print(x.shape)
        clsEmbedsExpand = self.classEmbedding
        clsEmbedsExpand = clsEmbedsExpand.expand(B,-1,-1)
        print('clsembeds',clsEmbedsExpand.shape) 
        x = paddle.concat([x,clsEmbedsExpand],axis = 1)
        # B,Hp*Wp+1,P*P*C ->B,197,768
        x = x + self.positionEmbedding
        # B,Hp*Wp+1,P*P*C ->B,197,768
        print(x.shape)
        
        x = self.linearProjection(x)
        # B,Hp*Wp,P*P*C ->B,196,768
        print(x.shape)
        return x



testX = image_list_224[0].astype(np.float32)/255.0
testX = testX[None,:,:,:]
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
#testX = paddle.randn((1,3,224,224))
vit_patch_embedding = PatchEmbedding(imageSize=224,inChannels=3,patchSize=16,embeddingDimension=768)
testX = vit_patch_embedding(testX)
print(testX.shape)
```

# MSA layer implementation
![msa.png](msa.png)

```python
class MultiHeadAttention(nn.Layer):
    def __init__(self,numHeads = 12, embeddingDimension = 768,attentionDropoutRate = 0):
        super(MultiHeadAttention,self).__init__()  
        self.numHeads = numHeads 
        self.embeddingDimension = embeddingDimension 
        self.attentionDropoutRate = attentionDropoutRate
        assert embeddingDimension %numHeads ==0,'embedding Dimension % numHeads must equal 0' + \
        'embedding dimension = {0},num heads ={1}'.format(self.embeddingDimension,self.numHeads)   
        self.headDimension = self.embeddingDimension//self.numHeads
        self.headDimensionScale = 1.0/(self.headDimension**0.5)
        print(f'head dimension scale {self.headDimensionScale}')

        # B,N,headDims
        self.queryProjection = nn.Linear(in_features=self.embeddingDimension,out_features = self.embeddingDimension)
        self.keyProjection = nn.Linear(in_features=self.embeddingDimension,out_features= self.embeddingDimension)
        self.valueProjection = nn.Linear(in_features=self.embeddingDimension,out_features=self.embeddingDimension)
        self.softmax = nn.Softmax(axis = -1)
        self.dropout = nn.Dropout(p = self.attentionDropoutRate )
        self.outProjection = nn.Linear(in_features= self.embeddingDimension,out_features= self.embeddingDimension)
        
    
    def forward(self,inputs):
        x = inputs 
        query = self.queryProjection(x)
        print(f'query shape {query.shape}')
        key  = self.keyProjection(x)
        print(f'key shape {key.shape}')
        value = self.valueProjection(x)
        print(f'value shape {value.shape}')
        B,N,D = query.shape
        query = query.reshape(shape = [B,N,self.numHeads,self.headDimension])
        query = query.transpose(perm = [0,2,1,3]) #B,headN,N,headDims
        print(f'query shape {query.shape}')
        key = key.reshape(shape = [B,N,self.numHeads,self.headDimension])
        key = key.transpose(perm = [0,2,1,3])     #B,headN,N,headDims
        print(f'key shape {key.shape}')
        key_T = key.transpose(perm = [0,1,3,2])
        print(f'key T shape {key_T.shape}')
        value = value.reshape(shape = [B,N,self.numHeads,self.headDimension])
        value = value.transpose(perm = [0,2,1,3])  #  B,headN,N,headDims
        print(f'value shape {value.shape}')
        attention = query @ key_T * self.headDimensionScale
        attention = self.softmax(attention)
        attention = self.dropout(attention)
        print(f'attention shape{attention.shape} ')
        x  = attention @ value #  B,headN,N,headDims
        x  = x.transpose(perm = [0,2,1,3]) # B,N,headN,headDims
        x = x.reshape(shape =[B,N,self.numHeads * self.headDimension]) # B,N,D 
        print(f'x shape{x.shape}')
        x = self.outProjection(x)
        print(f'out projection{x.shape}')
        return x 



testX = image_list_224[0].astype(np.float32)/255.0
testX = testX[None,:,:,:]
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
#testX = paddle.randn((1,3,224,224))
vit_patch_embedding = PatchEmbedding(imageSize=224,inChannels=3,patchSize=16,embeddingDimension=768)
testX = vit_patch_embedding(testX)
print(f'patch embedding shape{testX.shape}')

multi_head_attention = MultiHeadAttention(numHeads=12,embeddingDimension=768)
testX = multi_head_attention(testX)
print(f'multi head attention shape {testX.shape}')
print(f'model layers{multi_head_attention}')
```

# Transformer Encoder layer 
![transformer encoder layer](transformer_encoder.jpg)

```python
class TransformerEncoderLayer(nn.Layer):
    def __init__(self,
                embeddingDimension = 768,     
                numHeads = 12,                          
                mlpRatio =4,
                mlpDropoutRate = 0,
                attentionDropoutRate = 0,
                pathDropRate = 0,
                isTraining = False
                ):
        super(TransformerEncoderLayer,self).__init__() 
        self.embeddingDimension = embeddingDimension
        self.numHeads = numHeads
        self.mlpRatio = mlpRatio
        self.mlpDropoutRate = mlpDropoutRate
        self.attentionDropoutRate = attentionDropoutRate
        self.pathDropRate = pathDropRate
        self.isTraining = isTraining
        self.multiHeadAttention = MultiHeadAttention(numHeads = self.numHeads,\
                                                    embeddingDimension= self.embeddingDimension,\
                                                    attentionDropoutRate=self.attentionDropoutRate)  
        self.mlpLayer = MLP(inputDimension= self.embeddingDimension,\
                            outputDimension=self.embeddingDimension,\
                            mlpRatio=self.mlpRatio,\
                            dropoutRate=self.mlpDropoutRate)
        self.normForAttention = nn.LayerNorm(normalized_shape= self.embeddingDimension)
        self.normForMLP = nn.LayerNorm(normalized_shape= self.embeddingDimension)
        self.attentionPathDrop = PathDrop(dropPathRate= self.pathDropRate,\
                                        isTraining= self.isTraining)
        self.mlpPathDrop = PathDrop(dropPathRate= self.pathDropRate,\
                                    isTraining= self.isTraining)



    
    def forward(self,inputs):
        x = inputs
        res1 = x
        x = self.normForAttention(x)
        print(f'norm for attention shape{x.shape}')
        x = self.multiHeadAttention(x)
        print(f'multi head attention shape{x.shape}')
        x = self.attentionPathDrop(x) + res1
        res2 = x 
        x = self.normForMLP(x)
        x = self.mlpLayer(x)
        print(f'mlp layer shape{x.shape}')
        x = self.mlpPathDrop(x) + res2
        
        return x 




testX = image_list_224[0].astype(np.float32)/255.0
testX = testX[None,:,:,:]
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
#testX = paddle.randn((1,3,224,224))
vit_patch_embedding = PatchEmbedding(imageSize=224,inChannels=3,patchSize=16,embeddingDimension=768)
testX = vit_patch_embedding(testX)
transformer_encoder = TransformerEncoderLayer(embeddingDimension=768,\
                                            numHeads=12,mlpRatio=4,\
                                            mlpDropoutRate=0, attentionDropoutRate=0,\
                                            pathDropRate = 0.5, isTraining = True)
testX = transformer_encoder(testX)
print(f'patch embedding shape{testX.shape}')
print(f'transfomer encoder shape{testX.shape}')
```

# Path drop 

```python
class PathDrop(nn.Layer):
    def __init__(self,dropPathRate = 0,isTraining = True):
        super(PathDrop,self).__init__()
        self.dropPathRate = dropPathRate
        self.keepPathRate = 1- self.dropPathRate
        self.isTraining = isTraining




    def forward(self,inputs):
        x = inputs
        #  inputs: B,N,D
        B,N,D = inputs.shape
        if self.keepPathRate > 0  and self.isTraining == True: 

            randKeep = paddle.rand(shape = (B,),dtype = 'float32')  # B,
            maskKeep = randKeep < self.keepPathRate
            maskKeep = maskKeep[:,None,None]  # B,1,1
            maskKeep = maskKeep.astype(np.float32)
            
            x = x * maskKeep
            x = x / self.keepPathRate
            print(f'maskKeep {maskKeep}')
            print(f'x pathdrop {x}')
        return x 



testX0 = image_list_224[0].astype(np.float32)/255.0
testX1 = image_list_224[1].astype(np.float32)/255.0
testX2 = image_list_224[2].astype(np.float32)/255.0
testX0 = testX0[None,:,:,:]
testX1 = testX1[None,:,:,:]
testX2 = testX2[None,:,:,:]
testX =  np.concatenate([testX0,testX1,testX2],axis = 0)
print(testX.shape)
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
#testX = paddle.randn((1,3,224,224))
vit_patch_embedding = PatchEmbedding(imageSize=224,inChannels=3,patchSize=16,embeddingDimension=768)
testX = vit_patch_embedding(testX)
transformer_encoder = TransformerEncoderLayer(embeddingDimension=768,numHeads=12,mlpRatio=4,mlpDropoutRate=0, attentionDropoutRate=0)
testX = transformer_encoder(testX)
print(f'patch embedding shape{testX.shape}')
print(f'transfomer encoder shape{testX.shape}')
pathDrop  = PathDrop(dropPathRate = 0.6,isTraining= True )
testX = pathDrop(testX)
```

```python
class TransformerEncoder(nn.Layer):
    def __init__(self,
                numLayers = 12,
                embeddingDimension = 768,     
                numHeads = 12,                          
                mlpRatio =4,
                mlpDropoutRate = 0,
                attentionDropoutRate = 0,
                pathDropRate = 0,
                isTraining = False):
        super(TransformerEncoder,self).__init__()
        self.numLayers = numLayers
        self.embeddingDimension = embeddingDimension
        self.numHeads = numHeads
        self.mlpRatio = mlpRatio
        self.mlpDropoutRate = mlpDropoutRate
        self.attentionDropoutRate = attentionDropoutRate
        self.pathDropRate = pathDropRate
        self.isTraining = isTraining
        self.transformerEncoderLayers = nn.LayerList()
        for i in range(self.numLayers): 
            encoderLayer = TransformerEncoderLayer(embeddingDimension=self.embeddingDimension,
                                numHeads=self.numHeads,
                                mlpRatio=self.mlpRatio,
                                mlpDropoutRate=self.mlpDropoutRate,
                                attentionDropoutRate=self.attentionDropoutRate,
                                isTraining = self.isTraining)
            self.transformerEncoderLayers.append(encoderLayer)

    

    def forward(self,inputs):
        x = inputs 

        for i in range(self.numLayers):
            encoder = self.transformerEncoderLayers[i]
            x = encoder(x)
            print(f'encoder the {i} th layer ,x shape {x.shape}')


        return x


testX0 = image_list_224[0].astype(np.float32)/255.0
testX1 = image_list_224[1].astype(np.float32)/255.0
testX2 = image_list_224[2].astype(np.float32)/255.0
testX0 = testX0[None,:,:,:]
testX1 = testX1[None,:,:,:]
testX2 = testX2[None,:,:,:]
testX =  np.concatenate([testX0,testX1,testX2],axis = 0)
print(testX.shape)
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
#testX = paddle.randn((1,3,224,224))
vit_patch_embedding = PatchEmbedding(imageSize=224,inChannels=3,patchSize=16,embeddingDimension=768)
testX = vit_patch_embedding(testX)
transformer_encoder = TransformerEncoderLayer(embeddingDimension=768,numHeads=12,mlpRatio=4,mlpDropoutRate=0, attentionDropoutRate=0)
testX = transformer_encoder(testX)
print(f'patch embedding shape{testX.shape}')
print(f'transfomer encoder shape{testX.shape}')
pathDrop  = PathDrop(dropPathRate = 0.6,isTraining= True )
testX = pathDrop(testX)

transformerEncoder = TransformerEncoder(numLayers = 12,
                                        embeddingDimension = 768,     
                                        numHeads = 12,                          
                                        mlpRatio =4,
                                        mlpDropoutRate = 0.1,
                                        attentionDropoutRate = 0.1,
                                        pathDropRate = 0.3,
                                        isTraining = True)
testX = transformerEncoder(testX)                
```

# intergrate to Vision Transformer

```python
class VisionTransformer(nn.Layer):
    def __init__(self,
                imageSize = 224,
                inChannels = 3,
                patchSize = 16,
                embeddingDimension = 768,
                numLayers = 12,                     
                numHeads = 12,                          
                mlpRatio =4,
                mlpDropoutRate = 0,
                attentionDropoutRate = 0,
                pathDropRate = 0,
                isTraining = False):
        super(VisionTransformer,self).__init__()
        self.imageSize = imageSize
        self.inChannels = inChannels
        self.patchSize = patchSize
        self.embeddingDimension = embeddingDimension 
        self.numLayers  = numLayers 
        self.numHeads = numHeads
        self.mlpRatio = mlpRatio
        self.mlpDropoutRate = mlpDropoutRate
        self.attentionDropoutRate = attentionDropoutRate
        self.pathDropRate = pathDropRate
        self.isTraining = isTraining
        
        self.patchEmbedding = PatchEmbedding(self.imageSize,self.inChannels,\
                                            self.patchSize,self.embeddingDimension)
        self.vitEncoder = TransformerEncoder(self.numLayers,
                                            self.embeddingDimension,     
                                            self.numHeads,                          
                                            self.mlpRatio,
                                            self.mlpDropoutRate,
                                            self.attentionDropoutRate,
                                            self.pathDropRate,
                                            self.isTraining)
    

    def forward(self,inputs):
        x = inputs
        x = self.patchEmbedding(x)
        x = self.vitEncoder(x)

        return x 



testX0 = image_list_224[0].astype(np.float32)/255.0
testX1 = image_list_224[1].astype(np.float32)/255.0
testX2 = image_list_224[2].astype(np.float32)/255.0
testX0 = testX0[None,:,:,:]
testX1 = testX1[None,:,:,:]
testX2 = testX2[None,:,:,:]
testX =  np.concatenate([testX0,testX1,testX2],axis = 0)
print(testX.shape)
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)

vitTransformer = VisionTransformer(imageSize = 224,
                        inChannels = 3,
                        patchSize = 16,
                        embeddingDimension = 768,
                        numLayers = 12,                     
                        numHeads = 12,                          
                        mlpRatio =4,
                        mlpDropoutRate = 0.1,
                        attentionDropoutRate = 0.1,
                        pathDropRate = 0.5,
                        isTraining = True)
vitResult = vitTransformer(testX)
print(f'vit transformer shape {vitResult.shape}')
```

# ClassfierHead

```python
class ClassifierHead(nn.Layer):
    def __init__(self,
                inputDimension = 768,
                numClasses = 1000,                
                mlpRatio =4,
                dropoutRate = 0,
                isTraining = True,
                isSoftmax = True ):
        super(ClassifierHead,self).__init__()
        self.inputDimension = inputDimension
        self.numClasses = numClasses
        self.mlpRatio = mlpRatio
        self.dropoutRate = dropoutRate
        self.isTraining = isTraining
        self.isSoftmax = isSoftmax
        self.softmax = nn.Softmax(axis=-1)
        self.classifierHead = None
        
        if isTraining == True: 
            self.classifierHead =  self.Sequential(nn.LayerNorm(normalized_shape = self.inputDimension),                                    
                                    MLP(inputDimension = self.inputDimension,
                                    outputDimension= self.numClasses,
                                    mlpRatio= self.mlpRatio,
                                    dropoutRate= self.dropoutRate))

        if isTraining == False:
            self.classifierHead = nn.Sequential(nn.LayerNorm(normalized_shape=self.inputDimension),
                                                nn.Linear(in_features=self.inputDimension,\
                                                out_features=self.numClasses))
            


    
    def forward(self,inputs):
        x = inputs[:,0,:]
        
        x = self.classifierHead(x)
        if self.isSoftmax == True:
            x = self.softmax(x)
        

        return x 




testX0 = image_list_224[0].astype(np.float32)/255.0
testX1 = image_list_224[1].astype(np.float32)/255.0
testX2 = image_list_224[2].astype(np.float32)/255.0
testX0 = testX0[None,:,:,:]
testX1 = testX1[None,:,:,:]
testX2 = testX2[None,:,:,:]
testX =  np.concatenate([testX0,testX1,testX2],axis = 0)
print(testX.shape)
testX = testX.transpose(0,3,1,2)
testX = paddle.to_tensor(testX)
vitTransformer = VisionTransformer(imageSize = 224,
                        inChannels = 3,
                        patchSize = 16,
                        embeddingDimension = 768,
                        numLayers = 12,                     
                        numHeads = 12,                          
                        mlpRatio =4,
                        mlpDropoutRate = 0.1,
                        attentionDropoutRate = 0.1,
                        pathDropRate = 0.5,
                        isTraining = True)
vitResult = vitTransformer(testX)
print(f'vit transformer shape {vitResult.shape}')
classifierHead = ClassifierHead(inputDimension = 768,numClasses=1000,
                                mlpRatio=4,dropoutRate = 0.1,
                                isTraining = False,
                                isSoftmax=True)

testX = classifierHead(vitResult)
print(classifierHead)
print(testX.shape)
```
